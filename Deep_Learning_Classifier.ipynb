## Write a Python program to classify authentic and fake currency notes using learning classifier

```python
# To create a deep learning model

# Required imports
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Read data into a dataframe
df=pd.read_csv("/content/BankNote_Authentication.csv")
```

```python
# Extract Features X and target y
X=df.values[:,:-1]
y=df.values[:,-1]
```

```python
# Train test split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
# Initialize the deep neural network model with 7 layers, Sigmoid activation
model=Sequential()
model.add(Dense(5,input_shape=(X.shape[1],)))
model.add(Dense(1,activation='sigmoid'))
```

```python
# Fit the model on X_train and y_train for 10 epochs
model.compile(optimizer='adam',loss='binary_crossentropy')
model.fit(X_train,y_train,epochs=30,batch_size=32)
```

```python
# Make predictions using the model, this prediction is a floating point
y_pred=model.predict(X_test)
```

```python
# y_pred contains the probability in each row in X set
y_pred
```

```python
# The resulting array is flattened to one dimensions and converted to integer
y_pred=(y_pred>0.5).flatten().astype(int)
y_pred
```

```python
# Printing the accuracy of the model
print(accuracy_score(y_test,y_pred))
```
